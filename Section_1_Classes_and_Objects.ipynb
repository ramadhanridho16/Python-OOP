## Define class method

image.png

```python
# The following example defines a class method that returns an anonymous Person object:

class Person:
  counter = 0

  def __init__(self, name, age):
    self.name = name
    self.age = age
    Person.counter += 1

  def greet(self):
    return f"Hi, it's {self.name}."

  @classmethod
  def create_anonymous(cls):
    return Person('Anonymous', 22)
```

```python
# The following shows how to call the create_anonymous() class method:

anonymous = Person.create_anonymous()
print(anonymous.name) 
print(anonymous.age)
```

## Define static method

image.png

```python
# defines a class "TemperatureConverter" that has two static methods that convert from celsius to Fahrenheit and vice versa

class TemperatureConverter:
  @staticmethod
  def celsius_to_fahrenheit(c):
    return 9*c/5+32

  @staticmethod
  def fahrenheit_to_celsius(f):
    return 5*(f-32)/9
```

```python
# To call a static method, you use the ClassName.static_method_name() syntax
f = TemperatureConverter.celsius_to_fahrenheit(30)
print(f) 
```

```python
# you can define an Employee class that inherits from the Person class:

class Employee(Person):
  def __init__(self, name, age, job_title):
    super().__init__(name, age)
    self.job_title = job_title
```

```python
class Employee(Person):
  def __init__(self, name, age, job_title):
    super().__init__(name, age)
    self.job_title = job_title

  def greet(self):
    return super().greet() + f"I' an a {self.job_title}"
```

## Get the values of class variables

```python
# get the values of class variables, you use the dot notation (.)
print(HtmlDocument.extension)
print(HtmlDocument.version)
```

```python
# If you access a class variable that doesn't exist, you'll get an AttributeError exception
# HtmlDocument.media_type
```

```python
# Another way to get the value of a class variable is to use the getattr() function.
# getattr() function accepts an object and a variable name
# It returns the value of the class variable

extension = getattr(HtmlDocument, 'extension')
version = getattr(HtmlDocument, 'version')

print(extension)
print(version)
```
